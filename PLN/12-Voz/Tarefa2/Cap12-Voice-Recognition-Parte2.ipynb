### Audio Recorder

```python
# Imports

# Este pacote gera a gravação
import pyaudio

# Este pacote gera o arquivo no formato .wav
import wave
```

```python
# Nome do arquivo de saída
audio_file = "audio/saida.wav"
```

Chunk:

http://paulbourke.net/dataformats/audio/

```python
# Tamanho do chunk size de 1024 samples
chunk = 1024
```

```python
# Formato do sample
FORMAT = pyaudio.paInt16
```

```python
# Formato Mono. Mude para 2 se você quiser stereo
channels = 1
```

Sample Rate:
    
https://manual.audacityteam.org/man/sample_rates.html

```python
# 44100 samples por segundo
sample_rate = 44100
```

```python
# Tempo de gravação (5 segundos)
record_seconds = 5
```

```python
# Inicializa o objeto PyAudio 
p = pyaudio.PyAudio()
```

```python
# Abre o stream como entrada e saída
stream = p.open(format = FORMAT,
                channels = channels,
                rate = sample_rate,
                input = True,
                output = True,
                frames_per_buffer = chunk)
```

```python
# Lista para os frames
frames = []
```

```python
print("Iniciando a gravação. Diga alguma coisa.")

for i in range(int(44100 / chunk * record_seconds)):
    data = stream.read(chunk, exception_on_overflow = False)
    frames.append(data)

print("Gravação concluída.")
```

```python
# Encerra e fecha o stream
stream.stop_stream()
stream.close()
```

```python
# Finaliza o objeto PyAudio
p.terminate()
```

```python
# Visualiza dados de voz
frames
```

```python
# Abre o arquiivo em modo de escrita
wf = wave.open(audio_file, "wb")
```

```python
# Configura os canais
wf.setnchannels(channels)
```

```python
# Configura o formato do sample
wf.setsampwidth(p.get_sample_size(FORMAT))
```

```python
# Configura a taxa do sample
wf.setframerate(sample_rate)
```

```python
# Grava os frames no formato de bytes
wf.writeframes(b"".join(frames))
```

```python
# Encerra o arquivo
wf.close()
```
