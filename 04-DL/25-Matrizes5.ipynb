# Singular Value Decomposition

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# Versões dos pacotes usados neste jupyter notebook
%reload_ext watermark
%watermark -a "Data Science Academy" --iversions
```

```python
# Imagem Original
path = 'original.jpg'
img = Image.open(path)
s = float(os.path.getsize(path))/1000
print("Dimensões da Imagem: ",img.size)
plt.title("Imagem Original (%0.2f Kb):" %s)
plt.imshow(img)
```

```python
# Convertendo a imagem para escala de cinza usando Numpy
imggray = img.convert('LA')
imgmat = np.array( list(imggray.getdata(band = 0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure()
plt.imshow(imgmat, cmap = 'gray')
plt.title("Imagem depois de convertida no padrão de escala de cinza")
plt.show()
```

```python
# Compressão da Imagem
print("Após compressão: ")

# Aplicando SVD com Numpy
U, S, Vt = np.linalg.svd(imgmat) 

for i in range(5, 51, 20):
    cmpimg = np.matrix(U[:, :i]) * np.diag(S[:i]) * np.matrix(Vt[:i,:])
    plt.imshow(cmpimg, cmap = 'gray')
    title = " Imagem após =  %s" %i
    plt.title(title)
    plt.show()
    result = Image.fromarray((cmpimg ).astype(np.uint8))
result.save('imagem-comprimida.jpg')
```
